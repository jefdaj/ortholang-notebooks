# Evaluation order

You might be familiar with other langauages like Python, R, C, Bash, Ruby, Java, etc. Most of the popular ones are what's called "imperative" programming languages because you tell the computer a series of steps to perform in a specific order. OrthoLang is different! It's a "declarative" language where you define a bunch of variables in terms of each other, and then the computer figures out which order all the commands need to be run in to calculate your final result. Instead of saying what to do, you say what things are.

There are lots of really cool benefits to designing a language this way, but the first thing you'll probably notice is that you can't update a variable in a script. For example, this will sort of work in an interactive session:

```ortholang
n = 45
n = n + 5

:show
```

But in your final script, `n` can only have one value; assigning it twice like this is an error. It's like how you can't write `x = 45`, and then `x = 50` in the same set of math equations, because that's inconsistent.

Instead, try to think of the script as a static set of equations, and the interactive session as a dialog between you and the computer where you edit those equations and it checks that each new version is valid. Only the final valid equations are saved, and the `result` might be calculated in a totally different order than you followed when writing them.

It tends to make more sense if you look at it like a flowchart. Let's make a new script to illustrate:

```ortholang
:drop
```

```ortholang
var1 = 1.5
var2 = "nothing depends on var2 so far"
var3 = 2e-3 * var1
var4 = var3 * 5 - var1
result = var4

plot_vars "Order is based on dependencies:"
```

*Note: flowcharts are still an experimental feature. They break if you combine them with OrthoLang's "replace each" feature. But for simple scripts they're fine.*

When you assign a new variable, OrthoLang makes a note of which other variables it depends on. Then later when you ask it to calculate the value of any variable, it knows what else has to be calculated first. This script will calculate `var1`,  then `var3`, then `var4` and ignore `var2` because it's not used.
